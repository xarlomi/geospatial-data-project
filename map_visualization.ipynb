### 1.2 Maping

```python
ofi_spotting = folium.Map (location = [37.7736265,-122.4200582], zoom_start=13)


for i,row in mapa_data.iterrows():
    name = {"location" : [row["latitude"],row["longitude"]],
            "tooltip" : row["name"]}
    
    if row["category"] == "starbucks":
        icon = Icon(color = "darkred",
                    prefix = "fa",
                    icon = "coffee",
                    icon_color = "white")
    elif row["category"] == "nursery":
        icon = Icon(color = "pink",
                    prefix = "fa",
                    icon = "child",
                    icon_color = "lightblue")
    elif row["category"] == "offices":
        icon = Icon(color = "black",
                    prefix = "fa",
                    icon = "briefcase",
                    icon_color = "red")
    elif row["category"] == "sport":
        icon = Icon(color = "green",
                    prefix = "fa",
                    icon = "mountain",
                    icon_color = "white")
    elif row["category"] == "karaoke":
        icon = Icon(color = "orange",
                    prefix = "fa",
                    icon = "microphone",
                    icon_color = "white")

        
    Marker(**name,icon = icon).add_to(ofi_spotting)
```
